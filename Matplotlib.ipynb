1.1 'Generating Sin and Cos Graph using Matplotlib'


```python
x= np.linspace(0,10,100)

# Defining Wave
sin_y= np.sin(x)
cos_y= np.cos(x)

# Plotting the Wave
plt.plot(x,sin_y,label='sine')
plt.plot(x,cos_y,label='cos')

plt.legend(loc="lower left")
plt.title('sine & cosine wave')
plt.show()
```

1)Legend is used to add labels to the graph given in the plot function. The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as lower right, center, upper right, lower left, etc.
2)
title function adds a title to the plot

'-' will construct a line like in above plots

```python
plt.plot(x,sin_y,"-",label='sine')
plt.plot(x,cos_y,"*",label='cos')

plt.legend()
plt.show()
```

4. Different Types of Plots

4.1 Line Plot
A line plot is a type of plot used to visualize the relationship between two continuous variables. It is a plot that displays data points as connected line segments. Line plots are commonly used to visualize trends and patterns in data, especially when the data is continuous and has a large number of observations.

In Matplotlib, the plot function is used to create a line plot. You can specify the x and y values of the data points, along with other optional arguments such as color, line style, and markers . You can also add labels, titles, and legends to your line plot to provide context and clarity.

```python
plt.style.use('Solarize_Light2')

x= np.linspace(2,12,100)
y= np.cos(x)

plt.plot(x,y,label='cos')
plt.legend(loc="lower left")
plt.title('line plot')

plt.show()
```

4.1.2 Multiple Line Plots

```python
plt.style.use('Solarize_Light2')

x= np.linspace(2,12,100)
y1= np.sin(x)
y2= np.cos(x)

plt.plot(x,y1,'-',label='sin',color='red',linewidth=1)
plt.plot(x,y2,'*',label='cos',color='green',linewidth=1)


plt.legend(loc="lower left")
plt.title('Multiple line plot')

plt.show()
```

Can be done using plt.xlim for x axi,s

```python
x=np.linspace(0,10,200)

plt.xlim(0,10)
plt.ylim(-2.5,2.5)
y= np.sin(x)
plt.plot(x, y,label='sine')
plt.title('Sine Wave')
plt.legend()

plt.show()
```

4.1.6 Using .set to combine all of these commands

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y, label='sine')

# Using .set() method to set multiple properties at once
plt.gca().set(xlim=(0, 10), ylim=(-2.5, 2.5), title='Sine Wave')
plt.legend()

plt.show()
```

plt.gca() gets the current Axes instance.
.set() method is called on the Axes instance to set multiple properties at once, including xlim, ylim, and title.

4.2 Scatter Plots

A scatter plot is a type of data visualization in which individual data points are represented as points in a two-dimensional plane. The x-coordinate of each data point represents one feature of the data, and the y-coordinate represents another feature.

```python
n=500

x= np.random.randn(n)
y= np.random.randn(n)

plt.scatter(x,y,c="red",s= 30,alpha=0.8)

plt.title('scatter plot')

plt.show()
```

4.2.2 Multiple Scatter Plots

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, s = 10, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, s = 10, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

4.3 Histograms and Binnings

A histogram is a plot that displays the distribution of a set of continuous or discrete data. The data is divided into a set of intervals or bins, and the height of each bar in the plot represents the number of data points that fall within that bin. Binnings refers to the process of dividing the data into these bins. The choice of bin size can greatly affect the appearance of the histogram, so it's important to choose a bin size that accurately represents the underlying distribution of the data.

```python
data= np.random.randn(10000)

plt.hist(data,edgecolor='black')

plt.show()
```

```python
data= np.random.randn(10000)

plt.hist(data,bins=100,edgecolor='black', alpha = .8, color = 'red')

plt.title('random hostogram')
plt.xlabel='frequency'
plt.ylabel='values'
plt.show()
```

Individual Bins can be removed using histtype='stepfilled'

```python
data= np.random.randn(10000)

plt.hist(data,bins=100,edgecolor='black', alpha = .8, color = 'red',histtype='stepfilled')

plt.title('random hostogram')
plt.xlabel='frequency'
plt.ylabel='values'
plt.show()
```

4.3.2 Multiple histograms

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2
x3 = np.random.randn(1000) + 4

plt.hist(x1, bins = 20,  alpha = .8,edgecolor='black')
plt.hist(x2, bins = 20,  alpha = .8,edgecolor='black')
plt.hist(x3, bins = 20, alpha = .8,edgecolor='black')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

4.4 2D Histograms

plt.hist2d is a function in matplotlib that generates a 2D histogram. A 2D histogram is a representation of the distribution of data over two variables.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hist2d(x, y, bins=20, cmap='viridis')

plt.title('2D Histogram')
plt.xlabel('x')
plt.ylabel('y')

plt.show()

```

```python
plt.hist2d(x,y, bins = 20 , cmap = 'Blues')
plt.title('2D Histogram')
plt.ylabel('x')
plt.xlabel('y')
plt.colorbar()

plt.show()
```

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

plt.figure(figsize=(5,7))

plt.pie(data, labels= cars)

plt.show()
```

4.6 Multiple Subplots

Multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot

In matplotlib, you can create multiple subplots using the subplot function.

```python
x = np.linspace(0,10,500)

fig, ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sine')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('tan')

y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')

plt.tight_layout()

plt.show()
```

With Loops

```python
plt.subplots_adjust(hspace = .4, wspace = .4)

for i in range(1,7):
    
    plt.subplot(2,3,i)
    plt.text(0.5 , 0.5,str((2,3,i)),fontsize = 14,ha='center')
```

with loops

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```
